```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
heart_data=pd.read_csv("/content/heart.csv")
heart_data.head()
```

```python
x=heart_data.drop(columns="target",axis=1)
y=heart_data["target"]
x=np.asarray(x)
y=np.asarray(y)
```

Model Selection

1.comparing the models with default hyperparamater values using cross validation

```python
models=[LogisticRegression(max_iter=1000),SVC(kernel="linear"),KNeighborsClassifier(),RandomForestClassifier(random_state=0)]
```

```python
def compare_models_cross_validation():
  for model in models:
    cv_score=cross_val_score(model,x,y,cv=5)
    mean_acc=sum(cv_score)/len(cv_score)
    mean_acc=round(mean_acc*100,2)
    print("mean accuracy of ",model,"=",mean_acc,"%")
    print("_________________________________________")
```

```python
compare_models_cross_validation()
```

2.comparing the models with diff hyperparameters values using gridsearchcv

```python
model_list=[LogisticRegression(max_iter=10000),SVC(),KNeighborsClassifier(),RandomForestClassifier(random_state=0)]
```

```python
model_hyperparameters={
    "log_reg_hyper":{
        "C":[1,5,10,20]
    },
    "svc_hyper":{
        "kernel":['linear','poly','rbf','sigmoid'],
        'C':[1,5,10,20]
    },
    "KNN_hyper":{
        'n_neighbors':[3,5,10]
    },
    "random_hyper":{
        'n_estimators':[10,20,50,100]
    }
}
```

```python
model_keys=list(model_hyperparameters.keys())
```

```python
def modelselection(modellist,hyper):
  result=[]
  i=0
  for model in modellist:
    key=model_keys[i]
    params=hyper[key]
    i+=1
    classifier=GridSearchCV(model,params,cv=5)
    classifier.fit(x,y)
    result.append({'model_used':model,'highest_score':classifier.best_score_,'best_hyperparameters':classifier.best_params_})
  result_dataframe=pd.DataFrame(result,columns=['model_used','highest_score','best_hyperparameters'])
  return result_dataframe
```
